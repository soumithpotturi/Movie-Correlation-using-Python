```python
#Import Libraries 
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure
```

```python
df = pd.read_csv("movies.csv")
```

```python
df.head()
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
df.dtypes
```

```python
#changing the data type
df = df.fillna(0)
df['budget'] = df.budget.astype('int64')
df['gross'] = df.gross.astype('int64')
df['votes'] = df.votes.astype('int64')
```

```python
df.dtypes
```

```python
df.head()
```

```python
#There is a mismatch in the released date and year column

df['correctedYear'] = df['released'].astype(str).str[7:-15]
df.head(3)
```

```python
df = df.sort_values(by=['gross'], inplace = False, ascending = False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
df.drop_duplicates().head()
```

```python
#Correlation -  Higher the budget, higher the revenue
# Bigger companies bring in greater revnues
```

```python
# Scatter Plot Budget vs Gross Revnue

f, ax = plt.subplots(1)
plt.scatter(x=df['budget'], y = df['gross'])
plt.title('Budget Vs Gross Revnue')
plt.xlabel('Budget of the film')
plt.ylabel('Gross Revnue')
ax.set_ylim(bottom=0)
matplotlib.rc('figure', figsize=(10,8))
plt.show(f)
```

```python
df.head()
```

```python
# plot the budget vs gross using seaborn
#ax.set_ylim(bottom=0)
sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws={"color": "red"}, line_kws = {"color":"blue"});
```

```python
#Exact correlation percentage
df.corr(method = "pearson") #pearson #kendall #spearman
```

### High correlation exists between Budget and Gross Revenue - 75%.


```python
correlation_matrix = df.corr(method = "pearson")
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
matplotlib.rc('figure', figsize=(10,8))
plt.show();
```

```python
df_numerized = df
```

```python
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] =  df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method = "pearson")
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
matplotlib.rc('figure', figsize=(10,10))
plt.show();
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5 ]
high_corr
```

### Votes and the Budget of a film have the highest correlation with the Gross Revenue
### Bigger company didn't mean higher revenue.
